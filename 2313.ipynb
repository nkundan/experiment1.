```python
import numpy as np
import matplotlib.pyplot as plt

# Hard-coded variables
x_values = np.linspace(0, 50, 100)
plt.plot(x_values, 0.01 * x_values**2 - 0.2 * x_values + 5, label="Hard-coded")

# Simulated keyboard input
a, b, c = 0.02, -0.3, 6
plt.plot(x_values, a * x_values**2 + b * x_values + c, label="Keyboard Input")

# Simulated file data
coeffs = [(0.01, -0.2, 5), (0.015, -0.25, 5.5), (0.02, -0.3, 6)]
for i, (a, b, c) in enumerate(coeffs):
    plt.plot(x_values, a * x_values**2 + b * x_values + c, label=f"File Model {i+1}")

# Plot settings
plt.title("Weather Model: Quadratic Equation")
plt.xlabel("Time (days)")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid()
plt.show()
```
